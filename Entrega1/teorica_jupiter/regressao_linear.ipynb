```python
import numpy as np

# Matrizes
A = np.array([[1, 1],[1, 2],[1, 3]])
y = np.array([[1],[3],[2]])
#print(A.shape)

# Pseudo-inversa
pinv = np.linalg.pinv(A)
beta = pinv.dot(y)
print(beta)
```

```python
import matplotlib.pyplot as plt

# Gráfico
plt.figure(figsize=(6,4))
x = A[:,1]
plt.plot(x,y,'*b',label='Dados')
plt.plot(x,beta[1]*x+beta[0],'r',linewidth=2,label='Aproximação')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc=2, shadow=True)
plt.grid(True)
plt.title('Mínimos quadrados')
plt.show()
```
