```python
import numpy as np

#Matrizes
A=np.array([[1,1],[1,2],[1,3]])
b=np.array([[1],[3],[2]])

#Pseudo-inversor
pinv=np.linalg.pinv(A)
beta=pinv.dot(b)
print(beta)
```

```python
import matplotlib.pyplot as plt

#Gráfico
plt.figure(figsize=(6,4))
x=A[:,1]
plt.plot( x, b, 'b*', label='dados')
plt.plot( x, beta[0] + beta[1]*x, 'r', label='aproximação')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='best', shadow=True)
plt.grid(True)
plt.show()
```
